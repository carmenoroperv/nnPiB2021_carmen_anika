```r
library(ggplot2)
```

```r
data <- read.csv("data/Contour_plot/NN_bn_do.csv", sep = ";")
```

```r
head(data)
```

```r
str(data)
```

```r
data$Weight <- as.factor(data$Weight)
```

```r
p1 <- ggplot(data = data)+
    geom_point(aes(x=Hidden_size, y=Val_loss, colour = Weight), size = 4) + 
    xlab("Size of the hidden layer")+
    ylab("Validation loss")+
    ggtitle("Neural Net bn do")+
    theme_minimal()+
    theme(text = element_text(size = 20))  +
    theme(axis.text.x = element_text(size = 20))+
    theme(axis.text.y = element_text(size = 20))
          
    #geom_line(aes(x=Hidden_size, y=Val_loss))
```
